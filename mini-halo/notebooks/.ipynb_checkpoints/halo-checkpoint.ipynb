```python
from astropy.cosmology import FlatLambdaCDM
import astropy.units as u
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import HTML
from matplotlib.animation import FuncAnimation

# Constants
G = 4.30091e-6  # kpc * (km/s)^2 / Msun
c = 299792.458  # km/s

# Cosmology
cosmo = FlatLambdaCDM(H0=70, Om0=0.3)

# Lens and source redshifts
z_d = 0.0001
z_s = 0.001

# Concentration parameter and position x
c_vir = 16
x_pos = 1.0  # At r_s

# Lens masses in solar masses
lens_masses = [1e10, 1e4, 1e5]  # Msun

def delta_c(c):
    return (200/3) * c**3 / (np.log(1 + c) - c / (1 + c))

def rho_crit_z(z):
    Hz = cosmo.H(z).value  # Hubble parameter at redshift z in km/s/Mpc
    return 3 * (Hz**2) / (8 * np.pi * G)  # Msun/kpc^3

def r_vir(M, z):
    rho_c_z = rho_crit_z(z)
    return (3 * M / (4 * np.pi * 200 * rho_c_z))**(1/3)  # in kpc

def f_x(x):
    if np.isclose(x, 1.0):
        return 1/3
    elif x < 1:
        return (1 / (x**2 - 1)) * (1 - (2 / np.sqrt(1 - x**2)) * np.arctanh(np.sqrt((1 - x) / (1 + x))))
    else:
        return (1 / (x**2 - 1)) * (1 - (2 / np.sqrt(x**2 - 1)) * np.arctan(np.sqrt((x - 1) / (x + 1))))




# Store results
results = []

for M in lens_masses:
    rvir = r_vir(M, z_d)
    rs = rvir / c_vir
    dc = delta_c(c_vir)
    rho_c = rho_crit_z(z_d)
    
    rho_s = dc * rho_c

    
    kappa_s = rho_s * rs / Sigma_crit
    kappa_x = 2 * kappa_s * f_x(x_pos)
    results.append((M, kappa_s, kappa_x))



def KappaS(M,z_l,z_s):

    rvir = r_vir(M, z_l)
    c_vir =  concentration_duffy(M, z_l, h=0.7)
    rs = rvir / c_vir
    dc = delta_c(c_vir)
    rho_c = rho_crit_z(z_l)
     # Angular diameter distances in kpc
    Dd = cosmo.angular_diameter_distance(z_l).to(u.kpc).value
    Ds = cosmo.angular_diameter_distance(z_s).to(u.kpc).value
    Dds = cosmo.angular_diameter_distance_z1z2(z_l, z_s).to(u.kpc).value
    
    
    # Critical surface density (Msun / kpc^2)
    Sigma_crit = (c**2 / (4 * np.pi * G)) * (Ds / (Dd * Dds))
        
    rho_s = dc * rho_c
    kappa_s = rho_s * rs / Sigma_crit

    return kappa_s

```

```python
def concentration_duffy(M, z, h=0.7):
    """
    Compute the NFW concentration parameter using Duffy et al. (2008) for virial mass halos.
    
    Parameters:
    - M: Halo mass in solar masses (Msun)
    - z: Redshift
    - h: Dimensionless Hubble parameter (default: 0.7)

    Returns:
    - Concentration parameter c
    """
    # Duffy et al. 2008 parameters for virial mass
    A = 10.14
    B = -0.081
    C = -1.01
    M0 = 2e12 / h  # in Msun

    c = A * (M / M0)**B * (1 + z)**C
    return c

# Example usage:
halo_mass = 1e10  # Msun
redshift = 0.0001
c = concentration_duffy(halo_mass, redshift)
print(f"Concentration (c) for M = {halo_mass:.1e} Msun at z = {redshift}: {c:.2f}")
```

```python

def FermatPotential(x1,x2,y1,y2,psi,**kwargs):

    geo = 0.5*np.sqrt((x1-y1)**2+ (x2-y2)**2)
    grav =  psi(x1,x2,**kwargs)
    return geo- grav

def psi_NFW(x1, x2, ks):
    x = np.sqrt(x1**2 + x2**2)
    psi = np.zeros_like(x)

    # Masks
    mask_zero = x == 0
    mask_lt1 = (x > 0) & (x < 1)
    mask_ge1 = x >= 1

    # For 0 < x < 1
    psi[mask_lt1] = (ks / 2) * (
        (np.log(x[mask_lt1] / 2))**2
        - np.arctanh(np.sqrt(1 - x[mask_lt1]**2))**2
    )

    # For x >= 1
    psi[mask_ge1] = (ks / 2) * (
        (np.log(x[mask_ge1] / 2))**2
        + np.arctan(np.sqrt(x[mask_ge1]**2 - 1))**2
    )

    # x == 0 stays 0 in psi (already initialized)

    return psi

```

```python
y1, y2 = 0.01,0
print("Impact parameter y:",np.sqrt(y1**2 + y2**2))

x1 = np.linspace(-2,2,200)
x2 = np.linspace(-2,2,200)

X1,X2 = np.meshgrid(x1,x2)

FP = FermatPotential(X1,X2,y1,y2,psi_NFW,ks= 2.1)
plt.contourf(X1,X2,FP, 20)
plt.colorbar()
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
```

```python
ks_values = [1.0, 2.5, 5.0]
y1 = 0.7
y2 = 0.0
y = np.sqrt(y1**2 + y2**2)
# --- One figure, 3 subplots ---
fig, axes = plt.subplots(
    nrows=1, ncols=3,
    figsize=(20, 5),
    constrained_layout=True  # auto-adjust spacing
)
fig.suptitle(f"Fermat Potential for various $\kappa_s$ Values and Impact parameter y = {y} ", fontsize=16)

for ax, ks in zip(axes, ks_values):
    FP = FermatPotential(X1, X2, y1, y2, psi_NFW, ks=ks)
    cs = ax.contourf(X1, X2, FP, levels=20)
    ax.set_title(f"ks = {ks}")
    ax.set_xlabel("x1")
    ax.set_ylabel("x2")

# common colorbar
cbar = fig.colorbar(cs, ax=axes, orientation='vertical', fraction=0.02, pad=0.0)
cbar.set_label("Fermat Potential")
plt.savefig(fname = f"fermatpotential-y_{y}.png",dpi = 300)
plt.show()
```

### Animation

```python


# Grid setup
x1 = np.linspace(-2, 2, 300)
x2 = np.linspace(-2, 2, 300)
X1, X2 = np.meshgrid(x1, x2)

ks = 2.5
y2 = 0
y1_values = np.linspace(0, 1, 60)

fig, ax = plt.subplots(figsize=(6, 5))
levels = 30


# Initial plot to create the colorbar
FP = FermatPotential(X1, X2, y1_values[0], y2, psi_NFW, ks=ks)
cs = ax.contourf(X1, X2, FP, levels=levels, cmap='viridis')
ax.set_title(f"Fermat Potential (y1 = {y1_values[0]:.2f})")
ax.set_xlabel("x1")
ax.set_ylabel("x2")

# Create colorbar based on the initial plot
cbar = fig.colorbar(cs, ax=ax, orientation='vertical', fraction=0.02, pad=0.05)


def update(i):
    ax.clear()
    FP = FermatPotential(X1, X2, y1_values[i], y2, psi_NFW, ks=ks)
    ax.contourf(X1, X2, FP, levels=levels, cmap='viridis')

    ax.set_title(f"Fermat Potential (y1 = {y1_values[i]:.2f})")
    ax.set_xlabel("x1")
    ax.set_ylabel("x2")

anim = FuncAnimation(fig, update, frames=len(y1_values), interval=100, blit=False)

# This should now work!
HTML(anim.to_jshtml())

# Save the animation
anim.save('fermat_potential_animation.mp4', writer='ffmpeg', dpi= 300 ,fps=20)
```
