```python
interpolation_file = "outdir/interpolation/resultcw.hdf5"

with h5py.File(interpolation_file, 'r') as file:
    w_values = file['w_values'][:]
    y_values = file['y_values'][:]
    data = file['results'][:]  # (len(w), len(y), 2)

real_data = data[:, :,: ,0]
imag_data = data[:, :,:, 1]
```

```python
F = real_data[0,:,0] + imag_data[0,:,0]
F.shape
```
